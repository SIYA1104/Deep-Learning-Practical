# Student performance analysis and prediction using ANN model



```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 1. Data Collection & Preprocessing

```python
df = pd.read_csv(r"C:\Users\lenovo\Desktop\StudentPerformanceFactors.csv")
```

```python
df.head()
```

```python
# Select numeric columns only
X = df.select_dtypes(include=['number']).drop('Exam_Score', axis=1)

# Target
y = df['Exam_Score']

print(X.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Normalize
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Preprocessing Done вњ…")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae']
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2
)
```

```python
loss, mae = model.evaluate(X_test, y_test)

print("Test Loss (MSE):", loss)
print("Mean Absolute Error (MAE):", mae)
```

```python
predictions = model.predict(X_test)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": predictions.flatten()
})

print(results.head(10))
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predictions)
print("R2 Score:", r2)
```
